```python
import pandas as pd
import joblib
```

## Load the data in dataframe

```python
df = pd.read_csv("../data/credit_risk_dataset.csv")
df.head()
```

#### Removed all the null values

```python
target_data = df.pop('loan_status')
df['loan_status'] = target_data
df.isnull().sum()
```

```python
df['loan_int_rate'] = df['loan_int_rate'].fillna(df['loan_int_rate'].mean())
df['person_emp_length'] = df['person_emp_length'].fillna(df['person_emp_length'].mean())
df.isnull().sum().sum()
```

#### Label Encoding (All the text labels are changed to the numerical values)

(This encoding is saved in the model folder for future reference)

```python
from sklearn.preprocessing import LabelEncoder

# List of all the text-based columns
categorical_features = [
    'person_home_ownership', 
    'loan_intent', 
    'loan_grade', 
    'cb_person_default_on_file'
]
all_label_encoders = {}

df_processed = df.copy()
for col in categorical_features:
    le = LabelEncoder()
    df_processed[col] = le.fit_transform(df_processed[col])
    
    all_label_encoders[col] = le
    
joblib.dump(all_label_encoders, '../models/label_encoders.joblib')
```

```python
# saving processed data
df_processed.to_csv('../data/processed_data.csv', index=False)
```
